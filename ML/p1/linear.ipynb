```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn as sk 
```

```python
addr1 = "LR-Data1.csv"
addr2 = "LR-Data2.csv"
addr3 = "LR-Data3.csv"
```

```python
def Linear_reg(addr):
    data = pd.read_csv(addr,header=None).values
    m = data.shape[0]
    n = data.shape[1]
    x = x0 = data[:,0:-1]
    y = data[:,-1]
    I = np.ones([m,1])
    X = np.c_[x,I]

    Xt = np.transpose(X)
    XtX = np.dot(Xt,X)

    if np.linalg.det(XtX) == 0:
        XtX_inv = np.linalg.pinv(XtX)
    else:
        XtX_inv = np.linalg.inv(XtX)
        
    XtX_inv_X = np.dot(XtX_inv,Xt)
    W = np.dot(XtX_inv_X,y)

    return W
```

```python
a,b = Linear_reg(addr1)
print("回归方程：\n     y =",a,"*","x +",b)
data = pd.read_csv(addr1,header=None)
x0 = data[0]
y0 = data[1]
plt.plot(x0,y0,'.')
x = np.arange(min(x0)-0.1,max(x0)+0.1,0.001)
y = a*x+b
plt.plot(x,y)
```

```python
a,b = Linear_reg(addr2)
print("回归方程：\n     y =",a,"*","x +",b)
data = pd.read_csv(addr2,header=None)
x0 = data[0]
y0 = data[1]
plt.plot(x0,y0,'.')
x = np.arange(min(x0)-0.1,max(x0)+0.1,0.001)
y = a*x+b
plt.plot(x,y)
```

```python
a,b,c = Linear_reg(addr3)
print("分界线方程：\n   ",a,"*","x + ",b,"* y + ",c," = 0.5")
print("即：\n   y =",-a/b,"*","x + ",(0.5-c)/b)
data = pd.read_csv(addr3,header=None)
x0 = data[0]
y0 = data[1]
plt.plot(x0,y0,'.')
x = np.arange(min(x0)-0.1,max(x0)+0.1,0.001)
y = (-a/b)*x+(0.5-c)/b
plt.plot(x,y)
```
