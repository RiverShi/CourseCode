```python
def nofex(sstr):
    begin = 0
    if sstr[0] == '[' : begin = 1
    end = len(sstr)
    if sstr.find(']') == -1 : return sstr[begin:end]
    return sstr[begin:sstr.find(']')]
```

```python
times = {}

def add(x):
    if x not in times :
        times[x] = 1
    else : 
        times[x] = times[x] + 1

```

```python
text = open("199801.txt",'r',encoding="ANSI")
```

```python
while(True):
    
    str1 = text.readline()
    
    if str1 == "" :
        break
    if str1 == '\n' :
        continue
    
    str2 = str1.split("  ")[1:-1]
    str3 = list(map(nofex,str2))
    for x in str3 :
        if x[-2:] == '/w': continue
        add(x)
```
